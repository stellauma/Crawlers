```python
import requests
import pandas as pd
import math
```

```python
def source_name(source_dict):
    name = source_dict['name']
    return name
```

```python
keyword = 'malaria'
get_url = 'https://newsapi.org/v2/everything?'
apikey = 'cf22fe9246504b7f8d3e6e95a445458d'
page_count = 2
# main_df = pd.DataFrame()
```

```python
count=1
while count <= page_count:
    try:
        response = requests.get(f'{get_url}q={keyword}&from=2021-04-14&pageSize=100&page={count}&apiKey={apikey}')
        print(count)
        if count == 1:
            page_count = math.ceil(response.json()['totalResults']/100)
            articles = response.json()['articles']
            df = pd.DataFrame(articles)
            main_df = df.copy()
        else:
            articles = response.json()['articles']
            df = pd.DataFrame(articles)
            main_df.append(df)
    except Exception as ee:
        print('ee')
    count += 1
main_df['source'] =  main_df['source'].apply(lambda x: source_name(x))    
```
